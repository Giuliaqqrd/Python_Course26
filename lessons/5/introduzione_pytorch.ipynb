## 2. Cosa è un tensore?

Un **tensore** è la struttura dati centrale di PyTorch. Generalizza scalari, vettori e matrici a dimensioni arbitrarie:

| Dimensioni | Nome comune | Esempio |
|:---:|:---:|:---:|
| 0D | Scalare | `42` |
| 1D | Vettore | `[1, 2, 3]` |
| 2D | Matrice | `[[1,2],[3,4]]` |
| 3D+ | Tensore | immagine RGB, batch di dati |

> Concettualmente simile agli `ndarray` di NumPy, ma con supporto nativo per GPU e per la **differenziazione automatica**.

## 3. Creare tensori

```python
# --- Da dati esistenti ---
t_lista = torch.tensor([1, 2, 3, 4])
print("Da lista Python:    ", t_lista)

t_matrice = torch.tensor([[1.0, 2.0], [3.0, 4.0]])
print("Matrice 2x2:\n", t_matrice)
```

```python
# --- Tensori speciali ---
zeros   = torch.zeros(3, 4)        # tutto zeri
ones    = torch.ones(2, 3)         # tutto uni
eye     = torch.eye(3)             # matrice identità
random  = torch.rand(2, 2)         # valori casuali in [0, 1)
randn   = torch.randn(2, 2)        # distribuzione normale N(0,1)
arange  = torch.arange(0, 10, 2)   # come range()
linspace= torch.linspace(0, 1, 5)  # 5 valori equidistanti tra 0 e 1

print("zeros:\n",   zeros)
print("ones:\n",    ones)
print("eye:\n",     eye)
print("random:\n",  random)
print("arange:",    arange)
print("linspace:",  linspace)
```

## 4. Attributi di un tensore

Ogni tensore ha tre attributi fondamentali: **shape**, **dtype** e **device**.

```python
t = torch.randn(3, 4, 5)  # tensore 3D

print("Shape (dimensioni):", t.shape)       # torch.Size([3, 4, 5])
print("ndim  (n° di assi):", t.ndim)        # 3
print("dtype (tipo dato): ", t.dtype)        # torch.float32 (default)
print("device (dove vive):", t.device)       # cpu
print("Numero di elementi:", t.numel())      # 3 * 4 * 5 = 60
```

```python
# Specificare il dtype alla creazione
t_int   = torch.tensor([1, 2, 3], dtype=torch.int64)
t_float = torch.tensor([1, 2, 3], dtype=torch.float32)
t_bool  = torch.tensor([True, False, True], dtype=torch.bool)

print(t_int, t_int.dtype)
print(t_float, t_float.dtype)
print(t_bool, t_bool.dtype)

# Casting tra tipi
print(t_int.float())   # → float32
print(t_float.long())  # → int64
```

## 5. Operazioni sui tensori

```python
a = torch.tensor([[1.0, 2.0], [3.0, 4.0]])
b = torch.tensor([[5.0, 6.0], [7.0, 8.0]])

# Operazioni element-wise (elemento per elemento)
print("Somma:          \n", a + b)
print("Differenza:     \n", a - b)
print("Prodotto elem.: \n", a * b)   # NON prodotto matriciale
print("Divisione:      \n", a / b)
```

```python
# Prodotto matriciale
print("Matmul (@ ):     \n", a @ b)
print("torch.matmul:   \n", torch.matmul(a, b))

# Operazioni di riduzione
print("\nSomma totale:   ", a.sum())
print("Media:          ", a.mean())
print("Massimo:        ", a.max())
print("Argmax (indice):", a.argmax())

# Riduzione per asse
print("\nSomma per colonna (dim=0):", a.sum(dim=0))
print("Somma per riga    (dim=1):", a.sum(dim=1))
```

```python
# Broadcasting — PyTorch estende automaticamente le dimensioni compatibili
t = torch.ones(3, 3)
v = torch.tensor([1.0, 2.0, 3.0])  # shape [3]

print("Broadcast t + v:\n", t + v)
# v viene 'trasmesso' su ogni riga di t
```

## 6. Indicizzazione e slicing

La sintassi è la stessa di NumPy.

```python
t = torch.arange(12).reshape(3, 4)  # matrice 3x4
print("Tensore originale:\n", t)

print("\nPrima riga:          ", t[0])
print("Ultima riga:         ", t[-1])
print("Elemento [1, 2]:     ", t[1, 2])
print("Prime 2 righe:\n",       t[:2])
print("Colonne 1 e 2:\n",        t[:, 1:3])
print("Righe pari:\n",           t[::2])
```

```python
# Indicizzazione booleana (masking)
mask = t > 6
print("Maschera (t > 6):\n", mask)
print("Valori > 6:        ", t[mask])
```

## 7. Reshaping — cambiare la forma

```python
t = torch.arange(12)
print("Originale (1D):      ", t.shape)

t_2d = t.reshape(3, 4)
print("Reshape (3x4):\n",       t_2d)

t_3d = t.reshape(2, 2, 3)
print("\nReshape (2x2x3):\n",   t_3d)

# -1 = calcola automaticamente quella dimensione
t_auto = t.reshape(-1, 4)    # → shape [3, 4]
print("\nReshape (-1, 4):  ",   t_auto.shape)

# Flatten: riduce a 1D
print("Flatten:          ",   t_2d.flatten())
```

```python
# squeeze / unsqueeze — rimuovere o aggiungere dimensioni di size 1
t = torch.randn(1, 5, 1)
print("Originale shape:",   t.shape)            # [1, 5, 1]
print("Squeeze shape:  ",   t.squeeze().shape)  # [5]
print("Unsqueeze(0):   ",   t.squeeze().unsqueeze(0).shape)  # [1, 5]
```

## 8. Tensori e NumPy

PyTorch e NumPy convivono benissimo. **Attenzione**: i tensori CPU e gli array NumPy condividono la stessa memoria.

```python
# NumPy → PyTorch
arr = np.array([1.0, 2.0, 3.0])
t   = torch.from_numpy(arr)
print("NumPy → Tensor:", t)

# PyTorch → NumPy
t2  = torch.tensor([4.0, 5.0, 6.0])
arr2 = t2.numpy()
print("Tensor → NumPy:", arr2)

# Memoria condivisa: modificare uno modifica l'altro
arr[0] = 99
print("Dopo arr[0]=99, il tensore vale:", t)  # cambia anche t!
```

## 9. GPU: spostare i tensori su CUDA

Uno dei vantaggi principali di PyTorch è il supporto nativo alla GPU.

```python
# Sceglie GPU se disponibile, altrimenti CPU
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Dispositivo in uso: {device}")

t_cpu = torch.randn(3, 3)
t_gpu = t_cpu.to(device)          # sposta sulla GPU (o rimane su CPU)

print("Device del tensore:", t_gpu.device)

# Creare direttamente su GPU
t_direct = torch.ones(2, 2, device=device)
print("Tensore creato su:", t_direct.device)
```

> Per eseguire operazioni tra due tensori devono essere sullo **stesso device**.

## 10. Autograd — differenziazione automatica (preview)

PyTorch tiene traccia di tutte le operazioni su un tensore con `requires_grad=True`, permettendo di calcolare i **gradienti** automaticamente. Questo è il cuore dell'addestramento delle reti neurali.

```python
# Esempio: calcola dy/dx per y = x^2 + 3x + 1
x = torch.tensor(2.0, requires_grad=True)

y = x**2 + 3*x + 1
print(f"y = x² + 3x + 1 con x=2 → y = {y.item()}")

# Backpropagation: calcola dy/dx
y.backward()

# Gradiente: dy/dx = 2x + 3 → per x=2: 2*2+3 = 7
print(f"Gradiente dy/dx in x=2: {x.grad.item()}")
```

```python
# Disabilitare il tracking dei gradienti (utile a inference time)
with torch.no_grad():
    z = x * 2
    print("requires_grad dentro no_grad:", z.requires_grad)  # False
```
