# Pandas Tutorial
Pandas serve per lavorare con dati tabulari (fogli di calcolo, databse ecc). La libreria ci aiuta nella visualizzazione, pulizia e processamento dei dati.
I due oggetti core sono i `DataFrame` e le `Series`.

````python
import pandas as pd 
````

## DataFrame
Rappresentano delle tabelle e sono array di entries, ognuna delle quali ha i propri valori. Ogni entry corrisponde a una riga o una colonna.
Non si limitano a dati di tipo int ma possono includere anche stringhe. 

`pd.DataFrame` è il costruttore che genera gli oggetti di tipo DataFrame:
- sintassi: `{'nome_colonna' : [valori], ...}`
- è possibile assegnare anche delle etichette (labels) alle righe usando `index`.

`({'nome_colonna' : [valori], ...}, index = ['A','B'])`


````python
pd.DataFrame({'column1': [1, 2, 3], 'column2': [4, 5, 6]}, 
             index=['A', 'B', 'C'])
````

````python
pd.DataFrame({'Bob': ['I liked it.', 'It was awful.'], 
              'Sue': ['Pretty good.', 'Bland.']},
             index=['Product A', 'Product B'])
````

## Series
Le serie sono sequenze di valori sottoforma di lista. Rappresentano una singola colonna di un DataFrame a cui possiamo fornire un `name` e anche in questo caso possiamo assegnare delle etichette di riga con `index`.

````python
pd.Series([30, 35, 40], index=['2015 Sales', '2016 Sales', '2017 Sales'], name='Product A')
````

# Lavorare con dataset già esistenti
Nella maggior parte dei casi si fanno operazioni su dati già esistenti che possono essere trattati con Pandas. 
Possiamo scaricare i dataset automaticamente da Kaggle.

## Importare le librerie necessarie
```python
import os
import pandas as pd
import kagglehub
```

- `os` — gestisce percorsi e file in modo compatibile con tutti i sistemi operativi
- `pandas` — libreria per l'analisi e la manipolazione di dati tabellari
- `kagglehub` — permette di scaricare dataset direttamente da Kaggle senza farlo manualmente

## Scaricare il dataset e caricarlo in un DataFrame
```python
DATASET_ID = "zynicide/wine-reviews"
FILE_NAME  = "winemag-data-130k-v2.csv"
OUTPUT_DIR = os.path.join(os.getcwd(), "data", "wine")

kagglehub.dataset_download(DATASET_ID, output_dir=OUTPUT_DIR)

df = pd.read_csv(os.path.join(OUTPUT_DIR, FILE_NAME))
```

- `os.getcwd()` restituisce la cartella in cui si trova il notebook, così il path funziona su qualsiasi computer senza modifiche
- `os.path.join()` costruisce il percorso completo in modo sicuro su qualsiasi sistema operativo (Windows, Mac, Linux)
- Le costanti `DATASET_ID`, `FILE_NAME` e `OUTPUT_DIR` sono dichiarate in cima per renderle facili da modificare

````python
import os
import pandas as pd
import kagglehub

DATASET_ID = "zynicide/wine-reviews"
FILE_NAME  = "winemag-data-130k-v2.csv"
OUTPUT_DIR = os.path.join(os.getcwd(), "data", "wine") # "/cartella/del/notebook/data/wine"

kagglehub.dataset_download(DATASET_ID, output_dir=OUTPUT_DIR)

df = pd.read_csv(os.path.join(OUTPUT_DIR, FILE_NAME), index_col=0) # NB: il file ha una colonna "Unnamed: 0" che è un indice, quindi la usiamo come indice del DataFrame
````

## Operazioni comuni sui dati

Pandas offre molti strumenti per esplorare e trasformare un DataFrame. 
Di seguito le operazioni più utili che useremo durante il corso.

### Ordinamento
Possiamo ordinare il DataFrame in base ai valori di una o più colonne.
```python
df.sort_values("colonna")                          # ordine crescente
df.sort_values("colonna", ascending=False)         # ordine decrescente
df.sort_values(["colonna1", "colonna2"])           # più colonne
```

### Statistiche per colonna
Pandas permette di calcolare rapidamente statistiche di base su colonne numeriche.
```python
df["colonna"].mean()    # media
df["colonna"].min()     # minimo
df["colonna"].max()     # massimo
df["colonna"].median()  # mediana
```

### Contare valori unici
Utile per capire la distribuzione di una variabile categorica, ad esempio quanti vini provengono da ogni paese.
```python
df["colonna"].value_counts()  # frequenza di ogni valore
df["colonna"].nunique()       # numero di valori unici
df["colonna"].unique()        # lista dei valori unici
```

### Valori nulli
I dataset reali contengono spesso valori mancanti. Pandas offre strumenti per identificarli, rimuoverli o sostituirli.
```python
df.isnull().sum()                    # conta i nulli per colonna
df.dropna(subset=["colonna"])        # elimina righe con nulli in una colonna
df["colonna"].fillna(0)              # sostituisce i nulli con un valore fisso
df["colonna"].fillna(df["colonna"].mean())  # sostituisce con la media
```

### Aggiungere e rinominare colonne
Possiamo arricchire il DataFrame con nuove colonne calcolate o rinominare quelle esistenti.
```python
df["nuova_colonna"] = df["colonna1"] / df["colonna2"]   # nuova colonna calcolata
df.rename(columns={"vecchio_nome": "nuovo_nome"})        # rinomina
```

### Raggruppamento con groupby
Permette di aggregare i dati per categoria, ad esempio calcolare il punteggio medio per ogni paese.
```python
df.groupby("categoria")["colonna"].mean()                        # media per gruppo
df.groupby("categoria")["colonna"].agg(["mean", "min", "max"])   # più statistiche
```
